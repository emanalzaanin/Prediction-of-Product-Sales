##Clean Data




```python
# shape
df.shape
```

```python
df.dtypes
```

```python
# Check for duplicate
df.duplicated().sum()
```

```python
# Check for missing values
df.isnull().sum()
```

```python
# Fill missing values in the Item_Weight column with the mean value
df['Item_Weight'].fillna(df['Item_Weight'].mean(), inplace=True)
```

```python
# Fill missing values in the Outlet_Size column with the most frequent value (mode)
df['Outlet_Size'].fillna(df['Outlet_Size'].mode()[0], inplace=True)
```

```python
# Verify that missing values have been handled
df.isnull().sum()
```

```python
# Fix inconsistent category names in the Item_Fat_Content column
df['Item_Fat_Content'] = df['Item_Fat_Content'].replace({
    'LF': 'Low Fat',
    'low fat': 'Low Fat',
    'reg': 'Regular'
})
```
