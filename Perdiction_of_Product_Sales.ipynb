##Clean Data




```python
# shape
df.shape
```

```python
df.dtypes
```

```python
# Check for duplicate
df.duplicated().sum()
```

```python
# Check for missing values
df.isnull().sum()
```

```python
# Fill missing values in the Item_Weight column with the mean value
df['Item_Weight'].fillna(df['Item_Weight'].mean(), inplace=True)
```

```python
# Fill missing values in the Outlet_Size column with the most frequent value (mode)
df['Outlet_Size'].fillna(df['Outlet_Size'].mode()[0], inplace=True)
```

```python
# Verify that missing values have been handled
df.isnull().sum()
```

```python
# Fix inconsistent category names in the Item_Fat_Content column
df['Item_Fat_Content'] = df['Item_Fat_Content'].replace({
    'LF': 'Low Fat',
    'low fat': 'Low Fat',
    'reg': 'Regular'
})
```

```python
# Display the unique values in the Item_Fat_Content column
df['Item_Fat_Content'].value_counts()
```

```python
# Generate summary statistics for numerical columns
df.describe()
```

```python
# Print a summary of the dataset structure
print("Dataset Shape (Rows, Columns):", df.shape)
```

```python
# Print remaining missing values to confirm cleaning
print("\nMissing Values in Each Column:")
print(df.isnull().sum())
```

```python
# Print the data types of all columns
print("\nData Types of Each Column:")
print(df.dtypes)
```

#### Histogram of Numerical Features
This visualization shows the distribution of numerical variables in the dataset.

```python
ax = plt.figure(figsize=(10, 6))
sns.histplot(df['Item_Outlet_Sales'], kde=True)
plt.title('Distribution of Item Outlet Sales')
plt.xlabel('Sales')
plt.ylabel('Frequency')
plt.savefig('histogram_sales.png')
plt.close()

ax
```

#### Boxplots of Numerical Features
Boxplots help identify the spread of the data and detect potential outliers.

```python
ax = plt.figure(figsize=(10, 6))
sns.boxplot(x='Outlet_Type', y='Item_Outlet_Sales', data=df)
plt.title('Sales Distribution by Outlet Type')
plt.xticks(rotation=45)
plt.savefig('boxplot_sales_by_outlet.png')
plt.close()

ax
```

### Countplots
 - This countplot displays the frequency of each Item_Type in the dataset


```python
ax = plt.figure(figsize=(12, 6))
sns.countplot(y='Item_Type', data=df, order=df['Item_Type'].value_counts().index)
plt.title('Frequency of Item Types')
plt.savefig('countplot_item_type.png')
plt.close()

ax
```

### Correlation Heatmap of Numeric Features
 - This heatmap shows the correlation matrix of the numeric features in the dataset.

```python
ax = plt.figure(figsize=(10, 8))
numeric_df = df.select_dtypes(include=[np.number])
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap of Numeric Features')
plt.tight_layout()
plt.savefig('heatmap_correlation.png')
plt.close()
ax
```
